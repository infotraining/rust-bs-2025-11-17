# `Option<&mut T>` as field in struct

```rust
struct Data<'a> {
    data: Option<&'a mut Vec<i32>> 
}
```

```rust
impl<'a> Data<'a> {
    fn append(&mut self, value: i32)
    {
        match self.data {
            Some(ref mut data) => data.push(value),
            None => {}
        }

    }
}

fn main() {
    let mut data = vec![1, 2, 3];
    let mut d = Data { data: Some(&mut data) };
    d.append(4);

    assert_eq!(data, vec![1, 2, 3, 4]);
}
```
